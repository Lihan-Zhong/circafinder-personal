```python
def balanced_age_sampling(adata, age_col='age', random_seed=42):
    np.random.seed(random_seed)
    adata.obs['__age_group'] = adata.obs[age_col]
    
    # Find minimum group size
    min_cells = adata.obs['__age_group'].value_counts().min()
    print(f"Balanced sampling to {min_cells} cells per age group.")

    # Sample cells evenly from each age group
    sampled_indices = []
    for age in adata.obs['__age_group'].unique():
        age_group_cells = adata.obs[adata.obs['__age_group'] == age].index
        sampled_cells = np.random.choice(age_group_cells, min_cells, replace=False)
        sampled_indices.extend(sampled_cells)

    # Subset AnnData
    adata_balanced = adata[sampled_indices].copy()
    return adata_balanced

def adata_cluster_CV_age_balanced(adata, group_id_col='TOD', age_col='age', CV_folds=5, scaling=True):
    group_id = adata.obs[group_id_col].values
    age_group = adata.obs[age_col].values
    cell_names = adata.obs_names

    # Assign CV folds randomly
    sample_cv_group = np.array(list_assign_groups(range(len(adata.obs)), CV_folds))
    
    input_data = np.log1p(adata.X.copy())
    classifier = LinearSVC(random_state=42, max_iter=10000)
    
    truelab, pred, full_cell_name, age_labels = [], [], [], []
    
    for i in range(CV_folds):
        train_idx = sample_cv_group != (i + 1)
        test_idx = sample_cv_group == (i + 1)
        
        train, test = input_data[train_idx, :], input_data[test_idx, :]
        y_train, y_test = group_id[train_idx], group_id[test_idx]
        test_ages = age_group[test_idx]
        test_cell_names = cell_names[test_idx]

        if scaling:
            scaler = StandardScaler(with_mean=not issparse(train))
            train = scaler.fit_transform(train)
            test = scaler.transform(test)
        
        classifier.fit(train, y_train)
        predicted = classifier.predict(test)
        
        truelab.extend(y_test)
        pred.extend(predicted)
        full_cell_name.extend(test_cell_names)
        age_labels.extend(test_ages)
        
    result = pd.DataFrame({
        "Cell_name": full_cell_name,
        "TrueGroup": truelab,
        "PreGroup": pred,
        "Age": age_labels
    })
    return result
```

```python
import os
import sys
from pathlib import Path
import scanpy as sc
import numpy as np
import pandas as pd
import time as tm
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
import random
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.sparse import issparse

def adata_load(df_cell_file, df_gene_file, gene_count_file):
    adata = sc.read(gene_count_file).transpose()
    df_cell = pd.read_csv(df_cell_file)
    df_gene = pd.read_csv(df_gene_file)

    df_cell.index = df_cell["sample"]
    df_gene.index = df_gene["gene_id"]
    adata.obs = df_cell
    adata.var = df_gene
    return adata

def adata_sample_cells(adata, group_ID_list, n_cells):
    dfs = []
    adata.obs["__groups"] = group_ID_list
    for group_id in adata.obs["__groups"].unique():
        adata_group = adata[adata.obs['__groups'] == group_id]
        if len(adata_group) < n_cells:
            print(f'Warning: group {group_id} has less than {n_cells} cells. All cells will be included.')
            dfs.append(adata_group)
        else:
            sample_index = np.random.choice(adata_group.obs_names, size=n_cells, replace=False)
            adata_sample = adata_group[sample_index]
            dfs.append(adata_sample)
    adata_sampled = dfs[0].concatenate(dfs[1:])
    return adata_sampled

def list_assign_groups(sample_IDs, n):
    group_numbers = np.arange(1, n + 1)
    return np.random.choice(group_numbers, size=len(sample_IDs))

def shuffle_list(input_list):
    shuffled_list = input_list.copy()
    random.shuffle(shuffled_list)
    return shuffled_list

def adata_cluster_CV(adata, cell_names, group_id, CV_folds=5, scaling=True):
    """
    Perform cross-validation clustering using LinearSVC.
    
    Parameters:
      - adata: AnnData object with expression data.
      - cell_names: array-like of cell names.
      - group_id: array-like of true group labels.
      - CV_folds: Number of folds in cross-validation.
      - scaling: If True, apply StandardScaler to training/test data.
      
    Returns:
      - result: DataFrame with cell names, true labels, and predicted labels.
    """
    # Assign cells into CV folds randomly (ensuring splits across cells)
    sample_cv_group = np.array(list_assign_groups(list(range(len(adata.obs))), CV_folds))
    group_id = np.array(group_id)
    
    # Log-transform the data (common for count data)
    input_data = adata.X.copy()
    input_data = np.log1p(input_data)
    
    # Initialize the classifier
    Classifier = LinearSVC(random_state=42, max_iter=10000)
    truelab = []
    pred = []
    full_cell_name = []
    
    for i in range(CV_folds):
        train = input_data[sample_cv_group != (i + 1), :]
        test = input_data[sample_cv_group == (i + 1), :]
        
        y_train = list(group_id[sample_cv_group != (i + 1)])
        y_test = list(group_id[sample_cv_group == (i + 1)])
        test_cell_names = cell_names[sample_cv_group == (i + 1)]
        
        # Conditionally apply scaling if requested.
        if scaling:
            # Check if the data is sparse; if so, set with_mean=False.
            if issparse(train):
                scaler = StandardScaler(with_mean=False)
            else:
                scaler = StandardScaler()
            train = scaler.fit_transform(train)
            test = scaler.transform(test)
        
        # Fit and predict
        Classifier.fit(train, y_train)
        predicted = Classifier.predict(test)
        
        truelab.extend(y_test)
        pred.extend(predicted)
        full_cell_name.extend(test_cell_names)
        
    result = pd.DataFrame({"Cell_name": full_cell_name,
                           "TrueGroup": truelab,
                           "PreGroup": pred})
    return result

def adata_shuffle_cluster_CV(adata, cell_names, group_id, CV_folds=5, scaling=True):
    return adata_cluster_CV(adata, cell_names, shuffle_list(list(group_id)), CV_folds, scaling)

def pretty_confused(actual, predicted, fig_x=10, fig_y=10, text_scl=5):
    confusion = pd.crosstab(np.array(actual), np.array(predicted), rownames=['Actual'], colnames=['Predicted'])
    confusion_freq = confusion.copy()
    confusion = confusion.div(confusion.sum(axis=1), axis=0) * 100
    confusion_flatten = confusion.unstack().reset_index(name='Percent')
    confusion_flatten_freq = confusion_freq.unstack().reset_index(name='Freq')
    confusion_flatten["Freq"] = confusion_flatten_freq["Freq"]
    confusion_flatten['Label'] = np.round(confusion_flatten['Percent'], 0).astype(str) + "%\n\nn=" + confusion_flatten['Freq'].astype(str)
    plt.figure(figsize=(fig_x, fig_y))
    sns.heatmap(confusion, cmap='viridis', linewidths=0.1, annot=np.round(confusion).astype(str), fmt='', cbar=False)
    plt.show()
```
